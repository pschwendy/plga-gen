```python
# notebook for interpreting results from gen.cpp
import numpy as np
from matplotlib import pyplot as plt
```

```python
DOPs = np.arange(40, 1010, step=8)
```

```python
# plot mean L_L vs n from cpp generated values 
L_L_fixed_means = []
L_L_fixed_sems = []

L_L_unfixed_means = []
L_L_unfixed_sems = []

with open('data/L_L_means_f_d.txt', 'r') as f:
    L_L_fixed_means = [float(i.strip()) for i in f.readlines()]

with open('data/L_L_sems_f_d.txt', 'r') as f:
    L_L_fixed_sems = [float(i.strip()) for i in f.readlines()]

with open('data/L_L_means_d.txt', 'r') as f:
    L_L_unfixed_means = [float(i.strip()) for i in f.readlines()]

with open('data/L_L_sems_d.txt', 'r') as f:
    L_L_unfixed_sems = [float(i.strip()) for i in f.readlines()]

L_L_fixed_means = L_L_fixed_means[:122]
L_L_fixed_sems = L_L_fixed_sems[:122]
L_L_unfixed_means = L_L_unfixed_means[:122]
L_L_unfixed_sems = L_L_unfixed_sems[:122]

fig, ax = plt.subplots()
ax.errorbar(DOPs, L_L_fixed_means, yerr=np.array(L_L_fixed_sems), label='fixed')
ax.errorbar(DOPs, L_L_unfixed_means, yerr=np.array(L_L_unfixed_sems), label='unfixed')

# Expected L_L and L_L convergence as described in the paper
def E_L_L(n):
    return 8 * (n - 1) / (n - 2)
L_L_convergence = 8

ax.plot(DOPs, E_L_L(DOPs), label='$E_{L_L}$')
ax.axhline(L_L_convergence, linestyle='--', label='$L_L$ convergence')

ax.set_title("Mean $L_L$ for Different Degrees of Polymerization (dimers)")
ax.set_xlabel("Degree of Polymerization (n)", fontsize=12)
ax.set_ylabel("Mean $L_L$", fontsize=12)

ax.legend()
fig.savefig('data/L_L_dimers.png')
```
